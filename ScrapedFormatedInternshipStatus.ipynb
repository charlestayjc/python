```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import datetime
from bs4 import BeautifulSoup
from IPython.display import Image, HTML
from requests_html import HTMLSession # pip install requests-html
```

```python
URL = 'https://github.com/gcreddy42/hiring2020/blob/master/README.md#covid-19-summer-2020-internship-status-list'
page = requests.get(URL)
pageHTML = BeautifulSoup(page.content, 'html.parser')
```

```python
table = pageHTML.find_all('table')
df = pd.read_html(str(table))[0]
df['Logo'] = df['Company']
```

```python
def searchbingimg(SearchURL):
    r = HTMLSession().get(SearchURL)
    relement = r.html.find('img') # http://requests-html.kennethreitz.org/
    imagelink = []
    for element in relement:
        if element.search('alt'):
            imagelink.append(element)
    fullURL = str(imagelink[0])
    fullURL = fullURL[fullURL.find("https"):fullURL.find("width")-2]
    return fullURL

for companyname in df['Company']:
    try:
        companyindex = df[df['Company']==companyname].index.values.astype(int)[0]
        BOTWURL1 = 'https://www.brandsoftheworld.com/logo/'
        CompanyKeyword = companyname.replace(" ", "-", 1).replace("&", "").replace(" ", "")
        BOTWURL2 = '-0?original=1'
        SearchURL = BOTWURL1 + CompanyKeyword + BOTWURL2
        df['Logo'].iloc[companyindex]=str(searchbingimg(SearchURL))
    except Exception:
        pass
   
df['Logo'] = df['Logo'].apply(lambda x: '<img src="{0}" height="40" width="40">'.format(x))
df['Logo'] = df['Logo'].replace('<img src="https://www.brandsoftheworld.com/sites/all/themes/botw/img/logo-botw.png\' style" height="40" width="40">', '')
```

```python
df = df[df['Status'].notna()]
df = df.drop_duplicates(subset='Company', keep="last").sort_values(by=['Status','Company'])
df['Link'] ='http://' + df['Company'].astype(str) + ".com"
df['Link'] = df['Link'].apply(lambda x: '<a href="{0}">Link</a>'.format(x))
HTML(df.to_html(escape=False))
df.columns = ["Company", "Status", "Details", "Logo", "Link"]
df = df[['Logo', "Company", "Status", "Details", "Link"]]
df["Details"].fillna("No Additional Info", inplace = True) 
```

```python
# Colors elements in a dateframe green if not cancelled and red if cancelled. Else black.
def color_cancelled_red(value):
  if value == "Cancelled":
    color = 'red'
  elif value == "Not Cancelled":
    color = 'green'
  else:
    color = 'black'
  return 'color: %s' % color

# https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html
styles = [
    dict(selector="th", props=[("font-size", "115%"),
                               ("text-align", "left")]),
    dict(selector="td", props=[("font-size", "115%"),
                               ("text-align", "left")])
]

# https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html
Formated_Internship_Status = (df.style
    .applymap(color_cancelled_red, subset=['Status'])
    .hide_index()
    .set_table_styles(styles)
)
```
